```python
df1 = pd.get_dummies(df['Occupation'], drop_first = True)

```

```python
new_df = pd.concat([df1,df], axis=1)
new_df.drop(['ID', 'Occupation'],axis=1, inplace = True)
new_df.head()
```

```python
df2 = pd.get_dummies(new_df['Channel_Code'],drop_first = True)
new_df2 = pd.concat([df2,new_df], axis = 1)
new_df2.drop(['Channel_Code'], axis=1, inplace = True)
new_df2.drop(['Region_Code','Gender'], axis=1, inplace = True)
new_df2.head()
```

```python
df3 = pd.get_dummies(new_df2['Credit_Product'])
new_df3 = pd.concat([df3,new_df2], axis =1)
new_df3.drop(['Credit_Product'], axis=1, inplace = True)

new_df3.head()
```

```python
new_df3.rename(columns = {'No':'Credit_Product_No', 'Yes':'Credit_Product_Yes'}, inplace = True)
new_df3.columns
```

```python
df4 = pd.get_dummies(new_df3['Is_Active'])
final_df = pd.concat([df4,new_df3], axis =1)
final_df.drop(['Is_Active'], axis=1, inplace = True)
final_df.rename(columns = {'No':'Not_Activated', 'Yes':'Yes_Active'}, inplace = True)
final_df.head()

```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = final_df[['Vintage', 'Age', 'X3', 'X2', 'Credit_Product_Yes', 'Yes_Active', 'Self_Employed', 'Avg_Account_Balance', 'X4', 'Not_Activated', 'Salaried', 'Credit_Product_No']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif["Feature"] = variables.columns
vif
```

```python
final_df.drop(['Yes_Active','Not_Activated'], axis=1, inplace = True)
```

```python
X = final_df.drop(columns=['Is_Lead'])
y = final_df['Is_Lead']
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
def classify(model, X, y):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    # train the model
    model.fit(x_train, y_train)
    print("Accuracy:", model.score(x_test, y_test) * 100)
    
    # cross-validation
    score = cross_val_score(model, X, y, cv=5)
    print("CV Score:", np.mean(score)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model, X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model, X, y)
```
